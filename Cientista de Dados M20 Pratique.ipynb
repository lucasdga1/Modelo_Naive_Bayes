```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
import plotly.figure_factory as ff
```

Durante a aula nossa variável a ser prevista (churn) continha apenas 2 categorias, a base de vocês contém mais. O Naive Bayes pode ser aplicado para problemas de classificação com múltiplas classes da mesma forma que para problemas de classificação binária. O Naive Bayes é um algoritmo de classificação probabilístico que calcula a probabilidade de uma amostra pertencer a cada classe e seleciona a classe com a maior probabilidade como a previsão final.
Em resumo, o Naive Bayes pode ser aplicado da mesma maneira para problemas de classificação com múltiplas classes, e os mesmos princípios se aplicam em termos de treinamento, avaliação e aplicação do modelo.

```python
# Carregando bases de teste
X_test = pd.read_csv('Xp_test.csv')
y_test = pd.read_csv('Yp_test.csv')
X_train = pd.read_csv('xp_train_bal.csv')
y_train = pd.read_csv('yp_train_bal.csv')
```

# 2) Aplique o algoritmo de Naive Bayes aos dados de treinamento.

```python
# Como o scikit pede um array unidimensional vou gerar uma variável para isso
y1_train = y_train.values.ravel()
# Aplicando bayes aos dados de treinamento
naive_cred = GaussianNB()
naive_cred.fit(X_train, y1_train)
```

# 3) Faça a avaliação do modelo com os dados de treinamento.
Traga a acurácia, recall e plote a matriz de confusão. Não se esqueça de avaliar com suas palavras o desempenho do modelo, interpretando as métricas.

Dica: Para calcularmos o recall em classificação multi classe precisamos usar o atributo macro:
recall = recall_score(y_train, y_pred_train, average='macro')

```python
# Acurácia do modelo de treino
y_pred_train = naive_cred.predict(X_train)
acc = accuracy_score(y1_train, y_pred_train)
print(f'Acurácia: {acc}')
```

```python
# Fazendo o recall multi classe
recall = recall_score(y1_train, y_pred_train, average='macro')
print(f'Recall: {recall}')
```

```python
conf_matrix = confusion_matrix(y1_train, y_pred_train)
# Lista de classes de crédito
cred_class = ['Baixo', 'Médio', 'Alto']
# Plotando a matriz de confusão
fig = ff.create_annotated_heatmap(
    z=conf_matrix,
    x=cred_class,
    y=cred_class,
    colorscale='PuBuGn',
    showscale=True,
)
fig.update_layout(
    template = 'simple_white',
    title = 'Matriz de confusão',
    xaxis_title= 'Predicted labels',
    yaxis_title= 'True labels',

)
```

O modelo se saiu muito bem na previsão dos dados do treino.

Obteve uma acurácia de 98% e um recall de mesma intensidade. Isso pode se dar pelo fato das variáveis relacionadas com a classificação de crédito terem uma relação mais independente entre si no nosso dataframe.
